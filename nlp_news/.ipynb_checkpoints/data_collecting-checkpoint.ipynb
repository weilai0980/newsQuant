```python
# NLP utility functions

# TO DO:
# summarization via keyphrase extraction
# https://github.com/boudinfl/pke#implemented-models


def sent_token_proc(sent_tokens):
    
    '''
    Argu: tokens of a sentence
    Return: cleaned token without punctuation
    '''
    
    # punctuation
    symb_set = punctuation + '’'
    table = str.maketrans(symb_set, ' '*len(symb_set)) 
    
    tmp_tokens = [w.translate(table) for w in sent_tokens]

    # concatenation of word and punctuation, e.g. A women is walking, through the doors.A dog is following...
    tokens = []
    for t in tmp_tokens:
        for k in t.split():
            tokens.append(k)
    
    tokens = [i for i in tokens if i]
    
    # tokens that are not alphabetic
    # tokens = [word for word in tokens if word.isalpha()]
    
    # stop words
    # stop_words = set(stopwords.words('english'))
    # tokens = [w for w in tokens if not w in stop_words]
    
#     print('----', tokens)
    
    # text lemmatization
    wordnet_lemmatizer = WordNetLemmatizer()
    lemma_tokens = [ wordnet_lemmatizer.lemmatize(i) for i in tokens] 
    
    
    # digit token
    
    
    # short tokens
    # tokens = [word for word in tokens if len(word) > 1]
    
    # lower case token
    
    return lemma_tokens

# split by uppercase characters
def entity_token(s):
    
    '''
    Argu: entity
    Return: space seperated named entity 
    '''
    
    token_upper = re.sub( r"([A-Z])", r" \1", s).split()
    
    return ' '.join([i for i in token_upper])


def text_paragraph(text):
    
    '''
    Argu: raw text of a piece of news
    Return: concatenated paragraph of the news
    '''
    para = text.splitlines()
    concat_para = para[0].rstrip()
    
    for k in range(1, len(para)):
        
        if para[k]:
            if concat_para[-1] == '.':
                concat_para = concat_para + ' ' + para[k].rstrip()
            else:
                concat_para = concat_para + '. ' + para[k].rstrip()
                
    if concat_para[-1]!='.':
        concat_para = concat_para + '. '
        
    return concat_para

def text_sentences(text, title):
    
    '''
    Argu: text of a piece of news
    Return: list of sentences, top one is the title of the news 
    
    '''
    
    sentences = nltk.sent_tokenize( text )
    
    if title[-1] == '.':
        sentences.insert(0, title + ' ')
    elif title[-1] == '. ':
        sentences.insert(0, title)
    else:
        sentences.insert(0, title + '. ')
        
    sentences = [i for i in sentences if i]
        
    # filter out the footer including url or email
    urls = re.findall('https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+', sentences[-1])
    email = re.findall(r'[\w\.-]+@[\w\.-]+', sentences[-1])
        
    if urls or email:
        del sentences[-1]
        
    # filter out brakets 
    sentences= [tmp_sent for tmp_sent in sentences \
                if not(tmp_sent.find('(') == 0 and tmp_sent.find(')') == (len(tmp_sent)-2))]
    
    return sentences

def word_tokenize(sentence):
    
    return nltk.word_tokenize(sentence)

def NER_sentence(sentence):
    
    # Bi-LSTM model
    tmp_ent = model.analyze(sentence)['entities']
    
    # NLTK 
    '''
    pos = nltk.pos_tag(nltk.word_tokenize(sent))
            
    chunk = nltk.ne_chunk(pos)
            
    iob_chunk = nltk.chunk.tree2conlltags(chunk)
            
    cur_ck = ""
    cur_type = ""
    for ck in iob_chunk:
        
        if ck[2][0]=='B':
           cur_ck = ck[0]
           cur_type = ck[2][2:]
        
        elif ck[2][0]=='I':
           cur_ck = cur_ck + ' ' + ck[0]
        
        elif ck[2][0]=='O':
        # entity type, name
            if cur_ck:
                
                np_chunk.append([cur_type, cur_ck])
                cur_ck = ""
                cur_type = ""
    '''
    return tmp_ent

def sentence_entity_dict(entity_set, entity_sent_dict):
    
    ent_type = {}
    
    for k in entity_set:
        ent_type.update({k[1]:k[0]})
        
        
    sent_ent = {}
    for tmp_ent in entity_sent_dict:
        
        tmp_sent_list = entity_sent_dict[tmp_ent]
            
        for tmp_sent_id in tmp_sent_list:
            
            if tmp_sent_id not in sent_ent:
                sent_ent.update({tmp_sent_id:[]})
                sent_ent[tmp_sent_id].append( [tmp_ent, ent_type[tmp_ent]] )
                
            else:
                sent_ent[tmp_sent_id].append( [tmp_ent, ent_type[tmp_ent]] )
            
#     print(sent_ent, '\n\n', crypto_sent_dict)    
    
    return sent_ent


def entity_sentence_segment(ent_sent_list, sent_ent, max_num_sent):
    
    tmp_seg = []
    
    for tmp_id in ent_sent_list:
        
        # -- backward
        cur_id = tmp_id-1
        while cur_id >= 0:
            
            if cur_id in sent_ent:
                cur_sent_ent = sent_ent[cur_id]
                        
                for idx, tmp_ent in enumerate( cur_sent_ent ):
                    if tmp_ent[1] == 'ORG' or tmp_ent[1] == 'PER':
                        
                        break
                            
                if idx != len(cur_sent_ent):
                    break
                    
            cur_id -= 1
                
        left_bd = max(cur_id, 0)
                
        # -- forward
        cur_id = tmp_id+1
        while cur_id < max_num_sent:
            
            if cur_id in sent_ent: 
                cur_sent_ent = sent_ent[cur_id]
                        
                for idx, tmp_ent in enumerate( cur_sent_ent ):
                    if tmp_ent[1] == 'ORG' or tmp_ent[1] == 'PER':
                        break
                                
                if idx != len(cur_sent_ent):
                    break
                    
            cur_id += 1
                    
        right_bd = min(cur_id, num_sent-1)
                
        tmp_seg.append([left_bd, right_bd])
                
    # merge segments 
    cur_l = tmp_seg[0][0]
    cur_r = tmp_seg[0][1]
    merged_seg = []
            
    for tmp_l, tmp_r in tmp_seg:
        if (tmp_l-1) <= cur_r:
            cur_r = tmp_r
        else:
            merged_seg.append([cur_l, cur_r])
            cur_l = tmp_l
            cur_r = tmp_r
                    
    merged_seg.append([cur_l, cur_r])
    
    return merged_seg
             
    
# sentiment analysis: no need of casting, lower case     
def sentiment_analyzer(text):
    
    sid = SIA()
    
    return sid.polarity_scores(text)
```

```python
para_min_sentence = 5

local_content = pickle.load(open("./local_content.p", "rb")) 
print(' !!!  amount of news: ', len(local_content))

local_nlp = []
for news_idx, tmp in enumerate( local_content[40:80] ):

    tmp_text = tmp['text']
    
    # -- filtering news by entity
    
    #     for idx, ent in enumerate(tmp_entity):
        
    #         if ent in tmp_entity:
    #             break
            
    #     if len(tmp_entity) > 0 and idx == len(tmp_entity):
    #         print('! empty content !')
    
    # --- process each piece of news
    if tmp_text:
        
        print('\n -------- ', news_idx, '\n')
        
        # --- paragraphs

        concat_para = text_paragraph(tmp_text)
        
        # --- raw sentences
        
        sentences = text_sentences(concat_para, tmp['title'])
        
        # drop the last one in case of page footer
        del sentences[-1]
        
        # --- loop over each sentence
        
        # list of sentence tokens
        sent_token = []
        
        # initialize entity-sentence dictionary
        tmp_entity_set = set()
        
        for i in tmp['entity']:
            tmp_entity_set.add( (i[0], i[1]) )
            
        ent_sent_dict = {}
        crypto_sent_dict = {}
        
        for ent in tmp_entity_set:
            ent_sent_dict.update({ent[1]:set()})
        
        
        for idx, sent in enumerate(sentences):
            
            # --- senstence parsing
                
            if len(sent) < para_min_sentence:
                continue
            
            # word_tokenize
            tmp_tokens = word_tokenize(sent)
            
            # pre-process sentence tokens for NER 
            clean_token = sent_token_proc(tmp_tokens)
            clean_sent = ' '.join( [i for i in clean_token] )
            
            # store sentence tokens for materialization
            sent_token.append([])
            for t in tmp_tokens:
                
                if t not in '“' + '”' + "'s" + "''" + "``" + "..." + string.punctuation:
                    sent_token[-1].append( contractions[t] if t in contractions else t.lower())
            
                    
            # --- add entities
            
            print('-', idx, '-', clean_sent) 
            tmp_ent = NER_sentence(clean_sent) 
    
            for k in tmp_ent:
                tmp_entity_set.add( (k['type'], k['text']) )
                
            
            # entity location in sentences
            for ent in tmp_entity_set:
                
                if ent[1] in sent or ent[1] in sent:
                    
                    if ent[1] in ent_sent_dict:
                        ent_sent_dict[ent[1]].add(idx)
                    else:
                        ent_sent_dict.update({ent[1]:set([idx])})
            
            
            # cryptocurrency location in sentences
            low_tokens = nltk.word_tokenize(sent.lower())
            
            for ent in crypto_entity:
                
                # loop over all variants
                for k in ent:
                    
                    if k.lower() in low_tokens:
                        
                        if ent[0] in crypto_sent_dict:
                            crypto_sent_dict[ent[0]].add(idx)
                        else:
                            crypto_sent_dict.update({ent[0]:set([idx])})
            
        # --- crypto and generic entity cross check 
        
        tmp_ent_toremove = []
        
        for tmp_ent in ent_sent_dict:
            if tmp_ent in crypto_sent_dict:
                
                crypto_sent_dict[tmp_ent] = crypto_sent_dict[tmp_ent].union( ent_sent_dict[tmp_ent] )
                tmp_ent_toremove.append(tmp_ent)
        
        # remove crypto entities from general ones
        for tmp_key in tmp_ent_toremove:
            del ent_sent_dict[tmp_key]
            
            
        # --- entity link, disambigution, coreference    
        
        # disambiguate concatenated namt of entity
        # e.g.: Financial Times, FinancialTimes
        
        tmp_entity_name = [k[1] for k in tmp_entity_set]
        
        for tmp_ent in tmp_entity_name:
            
            ent_token = entity_token(tmp_ent)
            
            if ent_token != tmp_ent and ent_token in tmp_entity_name:
                
                ent_sent_dict[ent_token] = ent_sent_dict[ent_token].union( ent_sent_dict[tmp_ent] )
                del ent_sent_dict[tmp_ent]
                
 
        # --- sentence entity dictionary  
            
        sent_ent = sentence_entity_dict(tmp_entity_set, ent_sent_dict)
        
        # --- crypto entity specific sentence segments 
        
        num_sent = len(sentences)
        crypto_seg = []
        
        for tmp_crypto in crypto_sent_dict:
            
            tmp_crypto_sent_id = sorted( crypto_sent_dict[tmp_crypto] )
            
            # sentence segments
            merged_crypto_seg = entity_sentence_segment(tmp_crypto_sent_id, sent_ent, num_sent)
            print('\n ++ ', tmp_crypto, ' after : ', merged_crypto_seg)
            
            
            crypto_seg.append(merged_crypto_seg)
            
            
            # entity specific sentiment over each segment
            for sent_l, sent_r in merged_crypto_seg:
                
                tmp_seg_ent = set()
                
                for sent_idx in range(sent_l, sent_r+1):
                    
                    if sent_idx in sent_ent:
                        
                        for tmp_ent, tmp_type in sent_ent[sent_idx]:
                            tmp_seg_ent.add((tmp_ent, tmp_type))
                            
                
                tmp_seg_sent_merged = ' '.join( [sentences[sent_idx] for sent_idx in range(sent_l, sent_r+1)] )
                
                print('\n Seg ', sent_l, sent_r, ' : ', sentiment_analyzer(tmp_seg_sent_merged),\
                      '\n', tmp_seg_ent)
                
        print('\n')
        
        # --- print out sentences annotated by crypto entities
        for idx, sent in enumerate(sentences):
            
            tmp_flag = False
            for k in crypto_seg:
                
                for tmp_seg in k:
                    if idx>=tmp_seg[0] and idx<=tmp_seg[1]:
                        tmp_flag = True
                        break
                
                if tmp_flag == True:
                    break
            
            if tmp_flag == True:
                print("(", idx, ")", colored(sent, 'red'))
            else:
                print("(", idx, ")", colored(sent, ))
                    
            
        # --- information extracted for each news
        
        tmp_dict = {'title':tmp['title'], 'entity_sent':ent_sent_dict, \
                    'crypto_sent':crypto_sent_dict, 'sent':sentences, 'entity_type':tmp_entity_set }
        
        local_nlp.append(tmp_dict)
        
        print(crypto_sent_dict, '\n\n')
    
    # filiter out empty text 
    else:
        print('! empty content !')
```
